```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
random_state_value = 42

# Generate random data for clustering
np.random.seed(random_state_value)
data, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)
```

```python
# Function to perform K-means clustering and plot the results


def plot_kmeans_clusters(data, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state_value)
    labels = kmeans.fit_predict(data)

    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap="viridis", edgecolor="k", s=50)
    plt.scatter(
        kmeans.cluster_centers_[:, 0],
        kmeans.cluster_centers_[:, 1],
        c="red",
        marker="x",
        s=200,
        label="Centroids",
    )
    plt.title(f"K-means Clustering with {n_clusters} Clusters")
    plt.legend()
    plt.show()


plot_kmeans_clusters(data, n_clusters=2)
```

```python
# Function to plot the Elbow Curve to determine optimal clusters
def plot_elbow_curve(data, max_clusters=10):
    inertia = []  # To store the sum of squared distances for different cluster values

    # Loop over different values of k (number of clusters)
    for k in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=k, random_state=random_state_value)
        kmeans.fit(data)
        inertia.append(
            kmeans.inertia_
        )  # Inertia is the sum of squared distances to the centroids

    # Plotting the Elbow Curve
    plt.plot(range(1, max_clusters + 1), inertia, marker="o")
    plt.title("Elbow Method For Optimal Clusters")
    plt.xlabel("Number of Clusters")
    plt.ylabel("Inertia (Sum of Squared Distances)")
    plt.show()


# Plot the Elbow Curve to find the optimal number of clusters
plot_elbow_curve(data, max_clusters=10)
```

```python
# Function to perform K-means clustering and plot the results


def plot_kmeans_clusters(data, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state_value, n_init=10)
    labels = kmeans.fit_predict(data)

    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap="viridis", edgecolor="k", s=50)
    plt.scatter(
        kmeans.cluster_centers_[:, 0],
        kmeans.cluster_centers_[:, 1],
        c="red",
        marker="x",
        s=200,
        label="Centroids",
    )
    plt.title(f"K-means Clustering with {n_clusters} Clusters")
    plt.legend()
    plt.show()


# Loop through different cluster numbers from 2 to 10
for n_clusters in range(2, 11):
    plot_kmeans_clusters(data, n_clusters)
```

```python
# Generate random data for clustering (fixed random state for reproducibility)
np.random.seed(42)
data, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# List of different random states to test
random_states = [0, 21, 42, 84, 100]


# Function to perform K-means clustering and plot the results
def plot_kmeans_clusters(data, n_clusters, random_state):
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state, n_init=10)
    labels = kmeans.fit_predict(data)

    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap="viridis", edgecolor="k", s=50)
    plt.scatter(
        kmeans.cluster_centers_[:, 0],
        kmeans.cluster_centers_[:, 1],
        c="red",
        marker="x",
        s=200,
        label="Centroids",
    )
    plt.title(
        f"K-means Clustering (n_clusters={n_clusters}, random_state={random_state})"
    )
    plt.legend()
    plt.show()


# Test different random states with K-means clustering (fixed number of clusters = 3)
for state in random_states:
    plot_kmeans_clusters(data, n_clusters=3, random_state=state)
```

```python
# Set a random seed for reproducibility
random_state_value = 42

# 1. Generate spherical clusters (default assumption of K-means)
data_spherical, _ = make_blobs(
    n_samples=300, centers=3, cluster_std=1.0, random_state=random_state_value
)

# 2. Generate elongated (elliptical) clusters (violating K-means assumptions)
data_elongated = np.dot(np.random.randn(300, 2), np.random.randn(2, 2)) + np.array(
    [5, 5]
)


# Function to perform K-means clustering and plot the results
def plot_kmeans_clusters(data, n_clusters, title):
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state_value)
    labels = kmeans.fit_predict(data)

    # Plot the clustered data
    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap="viridis", edgecolor="k", s=50)
    plt.scatter(
        kmeans.cluster_centers_[:, 0],
        kmeans.cluster_centers_[:, 1],
        c="red",
        marker="x",
        s=200,
        label="Centroids",
    )
    plt.title(title)
    plt.legend()
    plt.show()


# Plot for spherical data (with 3 clusters)
plot_kmeans_clusters(
    data_spherical, n_clusters=3, title="K-means Clustering on Spherical Data"
)

# Plot for elongated data (with 3 clusters)
plot_kmeans_clusters(
    data_elongated, n_clusters=3, title="K-means Clustering on Elongated Data"
)
```

```python
n_samples = 1500
random_state = 170
transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]

X, y = make_blobs(n_samples=n_samples, random_state=random_state)
X_aniso = np.dot(X, transformation)  # Anisotropic blobs
X_varied, y_varied = make_blobs(
    n_samples=n_samples, cluster_std=[1.0, 2.5, 0.5], random_state=random_state
)  # Unequal variance
X_filtered = np.vstack(
    (X[y == 0][:500], X[y == 1][:100], X[y == 2][:10])
)  # Unevenly sized blobs
y_filtered = [0] * 500 + [1] * 100 + [2] * 10
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))

axs[0, 0].scatter(X[:, 0], X[:, 1], c=y)
axs[0, 0].set_title("Mixture of Gaussian Blobs")

axs[0, 1].scatter(X_aniso[:, 0], X_aniso[:, 1], c=y)
axs[0, 1].set_title("Anisotropically Distributed Blobs")

axs[1, 0].scatter(X_varied[:, 0], X_varied[:, 1], c=y_varied)
axs[1, 0].set_title("Unequal Variance")

axs[1, 1].scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_filtered)
axs[1, 1].set_title("Unevenly Sized Blobs")

plt.suptitle("Ground truth clusters").set_y(0.95)
plt.show()
```

```python
common_params = {"n_init": "auto", "random_state": random_state}
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))

y_pred = KMeans(n_clusters=2, **common_params).fit_predict(X)
axs[0, 0].scatter(X[:, 0], X[:, 1], c=y_pred)
axs[0, 0].set_title("Non-optimal Number of Clusters")

y_pred = KMeans(n_clusters=3, **common_params).fit_predict(X_aniso)
axs[0, 1].scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred)
axs[0, 1].set_title("Anisotropically Distributed Blobs")

y_pred = KMeans(n_clusters=3, **common_params).fit_predict(X_varied)
axs[1, 0].scatter(X_varied[:, 0], X_varied[:, 1], c=y_pred)
axs[1, 0].set_title("Unequal Variance")

y_pred = KMeans(n_clusters=3, **common_params).fit_predict(X_filtered)
axs[1, 1].scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_pred)
axs[1, 1].set_title("Unevenly Sized Blobs")

plt.suptitle("Unexpected KMeans clusters").set_y(0.95)
plt.show()
```

```python
y_pred = KMeans(n_clusters=3, **common_params).fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("Optimal Number of Clusters")
plt.show()
```

```python
y_pred = KMeans(n_clusters=3, n_init=10, random_state=random_state).fit_predict(
    X_filtered
)

plt.scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_pred)
plt.title("Unevenly Sized Blobs \nwith several initializations")
plt.show()
```

```python
from sklearn.mixture import GaussianMixture

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

y_pred = GaussianMixture(n_components=3).fit_predict(X_aniso)
ax1.scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred)
ax1.set_title("Anisotropically Distributed Blobs")

y_pred = GaussianMixture(n_components=3).fit_predict(X_varied)
ax2.scatter(X_varied[:, 0], X_varied[:, 1], c=y_pred)
ax2.set_title("Unequal Variance")

plt.suptitle("Gaussian mixture clusters").set_y(0.95)
plt.show()
```
