```python
import pandas as pd
```

```python
data = pd.read_csv("heart_attack_prediction_dataset.csv")
```

```python
pd.options.display.max_columns = None
```

```python
data.head()
```

```python
columns = ["Sex", "Diet", "Continent", "Hemisphere", "Country"]

for column in columns:
    data[column] = pd.Categorical(data[column]).codes
```

```python
data.head()
```

```python
columns = ["Exercise Hours Per Week", "Sedentary Hours Per Day", "BMI"]

for column in columns:
    data[column] = data[column].round().astype(int)
```

```python
data.head()
```

```python
data["Systolic Pressure"] = data["Blood Pressure"].str.split("/").str[0].astype(int)
data["Diastolic Pressure"] = data["Blood Pressure"].str.split("/").str[1].astype(int)
```

```python
data.head()
```

```python
data = data.drop(columns=["Patient ID", "Blood Pressure"])
```

```python
data.head()
```

```python
data["Income"] = ((data["Income"] // 1000) * 1000).astype(int)
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

columns_to_scale = [
    "Age",
    "Cholesterol",
    "Systolic Pressure",
    "Diastolic Pressure",
    "Exercise Hours Per Week",
    "Stress Level",
    "Sedentary Hours Per Day",
    "Income",
    "BMI",
    "Triglycerides",
    "Sleep Hours Per Day",
]

data[columns_to_scale] = scaler.fit_transform(data[columns_to_scale])
```

```python
X = data.drop("Heart Attack Risk", axis=1)
y = data["Heart Attack Risk"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)
```
